```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Scatterplot(Nunericl -Numerical)
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style=None,size='size')
```

```python
titanic=pd.read_csv('Titanic-Dataset.csv')
```

```python
#barplot (Numerical-Categorical)
sns.barplot(x='Pclass',y='Age',hue='Sex',data=titanic)
```

```python
#box Plot(NUmerical-Categorical)
sns.boxplot(x='Sex',y='Age',hue='Survived',data=titanic)
```

```python
#Distplot (Numerical- Categorical)
# sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False)
# sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
sns.kdeplot(data=titanic,x='Age',hue='Survived',fill=True,common_norm=False)
# common_norm if false it treat hue separately 


```

```python
#HeatMap(Categorical -Categorical)
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Sex').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
```

Binary target  mean = probability

Groupby + mean = conditional probability

Helps decide important features

```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```
